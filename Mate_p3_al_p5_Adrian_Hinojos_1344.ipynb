Problema 3: El triple de un número disminuido en 8 es igual a 22Paso 1: Análisis y resolución matemáticaIdentificar la incógnita: Sea $x$ el número que no conocemos.Traducción algebraica: * "El triple de un número" $\rightarrow 3x$"Disminuido en 8" $\rightarrow - 8$"Igual a 22" $\rightarrow = 22$Planteamiento de la ecuación:$$3x - 8 = 22$$Despeje de la variable: * Primero, movemos el 8 (que está restando) al otro lado sumando:$$3x = 22 + 8 \Rightarrow 3x = 30$$Luego, el 3 (que multiplica a la $x$) pasa dividiendo:$$x = \frac{30}{3}$$Resultado:
 $x = 10$
 Paso 2: Código en Python

```python
# Definimos los datos del problema
resultado_final = 22
valor_restado = 8
multiplicador = 3

# Paso 1: Sumamos el valor restado para cancelar la resta
paso_uno = resultado_final + valor_restado

# Paso 2: Dividimos para encontrar la incógnita
x = paso_uno / multiplicador

print(f"El número buscado en el Problema 3 es: {int(x)}")
```

Problema 4: La mitad de un número es igual a 15Paso 1: Análisis y resolución matemáticaIdentificar la incógnita: Sea $n$ el número desconocido.Traducción algebraica: * "La mitad de un número" $\rightarrow \frac{n}{2}$"Igual a 15" $\rightarrow = 15$Planteamiento de la ecuación:$$\frac{n}{2} = 15$$Despeje de la variable: * Para dejar sola a la $n$, el 2 que está dividiendo debe pasar al otro lado realizando la operación contraria (multiplicación):$$n = 15 \cdot 2$$Resultado: $n = 30$

Paso 2: Código en Python

```python
# Definimos el valor de la mitad
valor_mitad = 15
divisor = 2

# Despejamos multiplicando por 2
numero_original = valor_mitad * divisor

print(f"El número buscado en el Problema 4 es: {numero_original}")
```

Problema 5: Reparto de manzanas (Suma de tres partes)Supongamos el siguiente escenario típico: Ana, Beto y Carla se reparten 60 manzanas. Beto tiene el doble que Ana, y Carla tiene el triple que Ana.Paso 1: Análisis y resolución matemáticaDefinir las variables basándonos en la parte más pequeña (Ana):Ana: $x$Beto: $2x$ (doble)Carla: $3x$ (triple)Planteamiento de la ecuación: La suma de las tres partes es igual al total (60).$$x + 2x + 3x = 60$$Simplificar y despejar:Sumamos las "partes": $6x = 60$Dividimos el total entre el número de partes:$$x = \frac{60}{6} \Rightarrow x = 10$$Resultado final:Ana: 10 manzanas.Beto: $10 \cdot 2 = 20$ manzanas.

Carla: $10 \cdot 3 = 30$ manzanas.

```python
# Datos totales
total_manzanas = 60

# Definimos las proporciones (1 de Ana + 2 de Beto + 3 de Carla = 6)
total_partes = 1 + 2 + 3

# Calculamos el valor de una "parte" (lo que tiene Ana)
ana = total_manzanas / total_partes
beto = ana * 2
carla = ana * 3

print(f"Reparto: Ana ({int(ana)}), Beto ({int(beto)}), Carla ({int(carla)})")
```
